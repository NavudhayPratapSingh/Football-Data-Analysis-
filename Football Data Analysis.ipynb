```python
#import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
epl_df=pd.read_excel(r"C:\Users\ASUS\Downloads\EPL_20_21.ods", engine='odf')
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
epl_df.isnull().sum() # total no. of null values
```

```python
# create 2 new columns
epl_df['MinsPerMatch']=(epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch']=(epl_df['Goals']/epl_df['Matches']).astype(float)
epl_df.head()
```

```python
# Total Goals
tg=epl_df['Goals'].sum()
```

```python
#Penality Goals
Total_PenaltyGoals=epl_df['Penalty_Goals'].sum()
```

```python
#penality attempts
epl_df['Penalty_Attempted'].sum()
```

```python
# Pie Chart for Penalities missed vs Scored 
plt.figure(figsize=(13,6))
pl_not_scored=epl_df['Penalty_Attempted'].sum()-Total_PenaltyGoals
data=[pl_not_scored,Total_PenaltyGoals]
labels=['Penalties missed','Penalities Scored']
color=sns.color_palette('Set2')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```

```python
#unique position
epl_df['Position'].unique()
```

```python
#total fw players
epl_df[epl_df['Position']=='FW']
```

```python
#players from different nations
np.size(epl_df['Nationality'].unique())
```

```python
#Most players from Which countries
nationality=epl_df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar',figsize=(12,6),color=sns.color_palette("magma"))
```

```python
# clubs with maximum players in their squad
maxplayer=epl_df.groupby('Club')['Name'].count().sort_values(ascending=False).head(5)
maxplayer.plot(kind='bar',color=sns.color_palette("viridis"))
```

```python
# clubs with least players in their squad
minplayer=epl_df.groupby('Club')['Name'].count().sort_values(ascending=True).head(5)
minplayer.plot(kind='bar',color=sns.color_palette("viridis"))
```

```python
player_agegroup=epl_df.groupby('Age')['Name'].count().sort_values(ascending=False)
player_agegroup.plot(kind='bar',color=sns.color_palette('magma'))
plt.title('Player grouped by Age')
```

```python
# players based on age group
under20=epl_df[epl_df['Age']<=20]
age20_25=epl_df[(epl_df['Age']>20)&(epl_df['Age']<=25)]
age25_30=epl_df[(epl_df['Age']>25)&(epl_df['Age']<=30)]
above30=epl_df[epl_df['Age']>30]
x=np.array([under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above30['Name'].count()])
mylabels=["<=20",">20 & <=25",">25 & <=30",">30"]
plt.title('Total players with age',fontsize=20)
plt.pie(x,labels=mylabels,autopct="%.1f%%")
plt.show()
```

```python
# total under 20 players in each club
punder20=epl_df[epl_df['Age']<20]
punder20['Club'].value_counts().plot(kind='bar',color=sns.color_palette('cubehelix'))
plt.title('Total Player under 20 in each club')
plt.plot(punder20['Club'].value_counts())
```

```python
#under 20 in manU
punder20[punder20['Club']=='Manchester United']
```

```python
#average of each 
avgeachclubage=epl_df.groupby('Club')['Age'].mean().sort_values(ascending=True).plot(kind='bar')
```

```python
epl_df.groupby('Club')['Age'].mean().sort_values(ascending=False)
```

```python
# total assist from each club
assisteachclb=epl_df.groupby('Club')['Assists'].sum().sort_values(ascending=True).plot(kind='bar',color='Black')
plt.title('Plot of Club vs Total Assists')
```

```python
# Top 10 assits
epl_df.groupby('Name')['Assists'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
#most goals by players
top_10_goals=epl_df[['Name','Club','Goals','Matches']].nlargest(n=10,columns='Goals')
top_10_goals.plot(kind='bar',color=sns.color_palette('Set2'))
top_10_goals
```

```python
# Goals per match
goalspermatch=epl_df[['Name','Club','GoalsPerMatch']].nlargest(n=10,columns='GoalsPerMatch')
goalspermatch
```

```python
# pie chart - goals with assist and without assist
plt.figure(figsize=(14,7))
assists=epl_df['Assists'].sum()
data=[tg-assists,assists]
labels=['Goals without assists','Goals with assists']
color=sns.color_palette('Set1')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```
